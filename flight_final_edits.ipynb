```python
data['FlightDate'] = pd.to_datetime(data['FlightDate'])
data['CRSDepDateTime'] = pd.to_datetime(data['CRSDepDateTime'])
data['CRSArrDateTime'] = pd.to_datetime(data['CRSArrDateTime'])
```

```python
data['ArrTime'] = data['ArrTime'].astype(int)
data['ArrTime'] = data['ArrTime'].astype(str)
data['ArrTime'] = data['ArrTime'].str.zfill(4)
data['ArrTime'].head()
data['ArrTime'] = data['ArrTime'].astype(int)
```

```python
data['DepTime'] = data['DepTime'].astype(int)
data['DepTime'] = data['DepTime'].astype(str)
data['DepTime'] = data['DepTime'].str.zfill(4)
data['DepTime'].head()
data['DepTime'] = data['DepTime'].astype(int)
```

```python
data['DepDelay'] = data['DepDelay'].astype(int)
data['ArrDelay'] = data['ArrDelay'].astype(int)
data['ActualElapsedTime'] = data['ActualElapsedTime'].astype(int)
data['DepTime'] = data['DepTime'].astype(int)
```
